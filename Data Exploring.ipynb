```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
weatherfootdata = pd.read_csv("./data_files/FootTrafficWeather_July2022_Melbourne.csv")
```

```python
weatherfootdata
```

```python
plt.figure(figsize=(20,20))
for i, col in enumerate(weatherfootdata.columns):
    plt.subplot(4,5,i+1)
    plt.hist(weatherfootdata[col], alpha=0.3, color='b', density=True)
    plt.title(col)
    plt.xticks(rotation='vertical')
```

```python
g = sns.PairGrid(weatherfootdata, vars=['total_people','total_rain' , 'max_temp', 'min_temp', 
                             'solar_exp'])
g.map(sns.scatterplot)
plt.show()
```

```python
f, ax = plt.subplots(figsize=(11, 10))
corr = weatherfootdata.corr()
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(20, 220, n=200),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);
```
